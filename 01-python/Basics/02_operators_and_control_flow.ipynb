#**Operators**

*   Arithmetic Operators

*   Relational Operators
*   Logical Operators


*   Bitwise Operators

*   Assignment Operators
*   Membership Operators





```python
# Arithmetic Operators

print(5+6)
print(5-6)
print(5*6)
print(5/2)
print(5//2)
print(5%2)
print(5**2)
```

```python
# Relational Operators

print(4>5)
print(4<5)
print(4>=4)
print(4<=4)
print(4==4)
print(4!=4)
```

```python
# Logical Operators

print(1 and 0)

print(1 or 0)

print(not 1)
```

```python
# Bitwise Operators - operate on binary values

# bitwise and
print(2 & 3)

# bitwise or
print(2 | 3)

# bitwise xor
print(2 ^ 3)

# bitwise not
print(~3)

# bitwise leftshift
print(4 >> 2)

# bitwise rightshift
print(5 << 2)
```

```python
# Assignment Operators

# =
# a = 2
a = 2

# a = a+2
a += 2
print(a)
```

```python
# Membership Operators

# in/not in

print('D'not in 'Delhi')
print(1 in [2,3,4,5,6])
```

```python
# While loop example -> program to print the table
# Program -> Sum of all digits of a given number
# Program -> keep accepting numbers from users till he/she enters a 0 and then find the avg
```

```python
number = int(input('enter the number'))

i = 1         # i stands for iteration

while i<11:
  print(number*i)
  i += 1
```

```python
# while loop with else

x = 1

while x < 3:
  print(x)
  x += 1

else:
  print('limit crossed')
```

```python
curr_pop = 10000

for i in range(10,0,-1):
  print(i,curr_pop)
  curr_pop = curr_pop - 0.1*curr_pop
```
