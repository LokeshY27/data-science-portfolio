#Multiline Line statement
In Python, we can write a single logical statement across multiple physical lines using either implicit or explicit line continuation.

```python
# Explicit Line
 # use backslash at the very end of a line to tell
 # python that the statment continues on the next line.

a = 1 + 2 + 3 + \
    4 + 5 + 6 + \
    7 + 8
print(a)
```

```python
# Implicit Line
 # Another way to use parenthesis

a = (1 + 2 + 3 +
    4 + 5 + 6 +
    7 + 8)
print (a)
```

#Python Set
Set is an unordered collection of unique items. Set is defined by values separated by comma inside braces {}. Items in a set are not ordered. We can perform set operators like union, intersection on two sets. Set have unique values.

```python
a = {10, 30, 20, 40, 5}
print (a)
print (type(a))
```

```python
s = {10, 20, 20, 30, 30, 30}
print (s)                 # automatically set won't consider duplicate elements.
```

```python
print (s(1))      # we can't print particular element in set bcz its unordered
                  # collection of items.
```
