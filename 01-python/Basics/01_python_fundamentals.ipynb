#Multiline Line statement
In Python, we can write a single logical statement across multiple physical lines using either implicit or explicit line continuation.

```python
# Explicit Line
 # use backslash at the very end of a line to tell
 # python that the statment continues on the next line.

a = 1 + 2 + 3 + \
    4 + 5 + 6 + \
    7 + 8
print(a)
```

```python
# Implicit Line
 # Another way to use parenthesis

a = (1 + 2 + 3 +
    4 + 5 + 6 +
    7 + 8)
print (a)
```
